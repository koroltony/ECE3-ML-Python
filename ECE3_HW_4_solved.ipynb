a) Prove that $a$, $b$, and $c$ are linearly independent. **(4 points)**

\\

**Answer:** If $a$, $b$, and $c$ are linearly independent, then $\alpha a + \beta b + \gamma c = \vec{0}$ for real-valued scalars $\alpha, \beta, \gamma$ is **only** true if $\alpha = \beta = \gamma = 0$.

Since $a, b, c$ are vecors in 3-space, we have 3 equations in our system of equations:

$\begin{align}
\alpha \frac{\sqrt{3}}{3} + \beta \frac{\sqrt{3}}{3} + \gamma \frac{\sqrt{3}}{3} &= 0 \\
\alpha \frac{\sqrt{3}}{3} - \beta \frac{\sqrt{3}}{3} + \gamma \frac{\sqrt{3}}{3} &= 0 \\
\alpha \frac{\sqrt{3}}{3} + \beta \frac{\sqrt{3}}{3} - \gamma \frac{\sqrt{3}}{3} &= 0 \\
\end{align}$

From the first equation, $\alpha = -\beta - \gamma$. From the second equation, $\alpha = \beta - \gamma$. Combining these equations gives $-\beta - \gamma = \beta - \gamma \Rightarrow 2\beta = 0 \Rightarrow \beta = 0$ and $\alpha = -\gamma$. But from the third equation we have $\alpha = -\beta + \gamma = \gamma$ (since $\beta = 0$), so $\alpha = -\gamma = \gamma \Rightarrow \gamma = 0 \Rightarrow \alpha = 0$. 

Therefore $\alpha = \beta = \gamma = 0$ must be true for $\alpha a + \beta b + \gamma c = \vec{0}$ to be true, and $a$, $b$, and $c$ are linearly independent.

### Part 2: Functions with Vector Input and Output

Consider the function $\textbf{f}: \mathbb{R}^4 \rightarrow \mathbb{R}^2$ such that

f) Is the function $\textbf{f}: \mathbb{R}^4 \rightarrow \mathbb{R}^2$ linear? Is it affine? Explain why for both. **(4 points)**

\\

**Answer:** Consider $\alpha, \beta \in \mathbb{R}$ and $\textbf{a} = \begin{bmatrix}
a_1\\
a_2\\
a_3\\
a_4
\end{bmatrix}, \textbf{b} = \begin{bmatrix}
b_1\\
b_2\\
b_3\\
b_4
\end{bmatrix} \in \mathbb{R}^4$. Then

$\begin{align*}
\textbf{f}(\alpha \textbf{a} + \beta \textbf{b}) &= \textbf{f}\left(\begin{bmatrix}
\alpha a_1 + \beta b_1\\
\alpha a_2 + \beta b_2\\
\alpha a_3 + \beta b_3\\
\alpha a_4 + \beta b_4
\end{bmatrix}\right) = \begin{bmatrix}
\alpha a_1 + \beta b_1 + \alpha a_2 + \beta b_2 \\
\alpha a_3 + \beta b_3 - \alpha a_4 - \beta b_4
\end{bmatrix} \\
&= \begin{bmatrix}
\alpha (a_1 + a_2) + \beta (b_1 + b_2) \\
\alpha (a_3 - a_4) + \beta (b_3 - b_4)
\end{bmatrix} \\
&= \begin{bmatrix}
\alpha (a_1 + a_2)\\
\alpha (a_3 - a_4)
\end{bmatrix} + \begin{bmatrix}
\beta (b_1 + b_2) \\
\beta (b_3 - b_4)
\end{bmatrix} \\
&= \alpha \begin{bmatrix}
a_1 + a_2\\
a_3 - a_4
\end{bmatrix} + \beta \begin{bmatrix}
b_1 + b_2 \\
b_3 - b_4
\end{bmatrix} \\
\alpha \textbf{f}(\textbf{a}) &= \alpha \textbf{f}\left(\begin{bmatrix}
a_1\\
a_2\\
a_3\\
a_4
\end{bmatrix}\right) = \alpha \begin{bmatrix}
a_1 + a_2\\
a_3 - a_4
\end{bmatrix} \\
\beta \textbf{f}(\textbf{b}) &= \beta \textbf{f}\left(\begin{bmatrix}
b_1\\
b_2\\
b_3\\
b_4
\end{bmatrix}\right) = \beta \begin{bmatrix}
b_1 + b_2\\
b_3 - b_4
\end{bmatrix}
\end{align*}$

\\

Therefore it is shown that $\textbf{f}(\alpha \textbf{a} + \beta \textbf{b}) = \alpha \textbf{f}(\textbf{a}) + \beta \textbf{f}(\textbf{b})$ for any $\alpha, \beta \in \mathbb{R}$ and $\textbf{a}, \textbf{b}\in \mathbb{R}^4$. $\textbf{f}: \mathbb{R}^4 \rightarrow \mathbb{R}^2$ satisfies superposition and is **linear**. Since this includes $\alpha, \beta\in\mathbb{R}$ such that $\alpha + \beta = 1$, $\textbf{f}: \mathbb{R}^4 \rightarrow \mathbb{R}^2$ is **affine**.

## Problem 3 (4 points)

Complete the following code to get the desired output as shown below:

![plot](https://raw.githubusercontent.com/monsij/ToDo_App/master/image.png)

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

n_rays = 50
x = np.array([0,1])
figure(figsize=(5,5))

for i in range(n_rays):
    theta = 2 * i * np.pi / n_rays
    Rtheta = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])
    y = Rtheta @ x
    plt.arrow(0, 0, y[0], y[1], color=f"C{i}")
plt.xlim([-1,1])
plt.ylim([-1,1])
#plt.axis('off');
plt.show()
```
