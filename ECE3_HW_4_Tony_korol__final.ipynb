## Problem 3 (4 points)

Complete the following code to get the desired output as shown below:

![plot](https://raw.githubusercontent.com/monsij/ToDo_App/master/image.png)

```python
import matplotlib.pyplot as plt
import numpy as np

n_rays = 50
x = np.array([0, 1])

for i in range(n_rays):
    theta = 2 * i * np.pi / n_rays
    Rtheta = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
    y =np.inner(Rtheta,x.T)
    plt.arrow(0, 0, 0.035*y[0],0.035*y[1], width=0.0001, color=f"C{i}")
plt.axis('equal')
plt.axis('off');
print(Rtheta)
```

```python
# Alternative solution

import matplotlib.pyplot as plt
import numpy as np

n_rays = 50
#x = np.array([0, 1])
for i in range(n_rays):
    theta = 2 * i * np.pi / n_rays
    Rtheta = theta+3*(2*np.pi/n_rays)
    y = np.sin(Rtheta)
    x = np.cos(Rtheta)
    plt.arrow(0, 0, 0.035*x, 0.035*y, width=0.0001, color=f"C{i}")
plt.axis('equal')
plt.axis('off');
```
